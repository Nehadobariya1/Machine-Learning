```python
#IMAGE FILTERING

import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
img=cv2.imread("nature2.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
plt.axis("off")
```

```python
#APPLYING KERNAL TO IMAGE(2D FILTERING)
plt.figure(figsize=(5,5))
# k1=np.array([[0,0,0],[0,1,0],[0,0,0]])
k1=np.array([[-1,-1,-1],[-1,8,0-1],[-1,-1,-1]])
filtered_img=cv2.filter2D(img,-1,k1)
plt.imshow(filtered_img)
plt.axis("off")
```

```python
#BLUR AN IMAGE
plt.figure(figsize=(5,5))
##blur(image,(x,y))
blur_img=cv2.blur(new_img,(15,15))
plt.imshow(blur_img)
plt.axis("off")
```

# Box-Blur

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("nature2.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,10))
blur_img1=cv2.blur(new_img,(15,15))
plt.subplot(1,3,1),plt.imshow(blur_img1),plt.title("Box-Blur"),plt.axis("off")

blur_img2=cv2.GaussianBlur(new_img,(5,5),0)
plt.subplot(1,3,2),plt.imshow(blur_img2),plt.title("Gaussian-Blur"),plt.axis("off")

blur_im3=cv2.medianBlur(new_img,3)
plt.subplot(1,3,3),plt.imshow(blur_img2),plt.title("median-Blur"),plt.axis("off")
```

# Blur Using Kernal

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("nature2.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(7,7))

k1=np.ones((8,8))/64
# print(k1)
blurred_img = cv2.filter2D(new_img,-1,k1)
plt.subplot(1,2,1)
plt.imshow(blurred_img)
plt.axis("off")

k1=np.zeros((8,8))/64
# print(k1)
blurred_img = cv2.filter2D(new_img,-1,k1)
plt.subplot(1,2,2)
plt.imshow(blurred_img)
plt.axis("off")
```

# Sharpening

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("nature2.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))

k1=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
blurred_img1 = cv2.filter2D(new_img,-1,k1)
plt.subplot(1,2,1)
plt.imshow(blurred_img1)
plt.axis("off")

k2=np.array([[1,1,1],[1,-7,1],[1,1,1]])
blurred_img2 = cv2.filter2D(new_img,-1,k2)
plt.subplot(1,2,2)
plt.imshow(blurred_img2)
plt.axis("off")
```

# Embossing (Shadow)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("nature2.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))

k1=np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
blurred_img1 = cv2.filter2D(new_img,-1,k1)+128
plt.subplot(1,4,1),plt.imshow(blurred_img1),plt.axis("off"),plt.title("South-East")

k2=np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
blurred_img2 = cv2.filter2D(new_img,-1,k2)+128
plt.subplot(1,4,2),plt.imshow(blurred_img2),plt.axis("off"),plt.title("South-west")

k3=np.array([[0,0,1],[0,0,0],[-1,0,0]])
blurred_img3 = cv2.filter2D(new_img,-1,k3)+128
plt.subplot(1,4,3),plt.imshow(blurred_img3),plt.axis("off"),plt.title("North-East")

k4=np.array([[1,0,0],[0,0,0],[0,0,-1]])
blurred_img4 = cv2.filter2D(new_img,-1,k4)+128
plt.subplot(1,4,4),plt.imshow(blurred_img4),plt.axis("off"),plt.title("North-west")
```
